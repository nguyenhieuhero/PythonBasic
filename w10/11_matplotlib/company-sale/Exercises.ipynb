# Visualizing Company Sale data

### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://github.com/thieu1995/csv-files/blob/main/data/matplotlib/company_sales_data.csv).

```python
data = pd.read_csv("https://raw.githubusercontent.com/thieu1995/csv-files/main/data/matplotlib/company_sales_data.csv")
data.head()
```

### Exercise 1: Read Total profit of all months and show it using a line plot

![img](img/01.png)

```python
plt.plot(data['month_number'],data['total_profit'])
plt.xlabel("Month")
plt.ylabel("Total profit")
```

### Exercise 2: Get total profit of all months and show line plot with the following Style properties

![img](img/02.png)

```python
plt.plot(data['month_number'],data['total_profit'],linestyle='dashed',marker='o',color='red',mfc='black',label='Profit data of last year')
plt.xlabel("Month")
plt.ylabel("Total profit")
plt.legend()
```

### Exercise 3: Read all product sales data and show it  using a multiline plot

![img](img/03.png)

```python
labels = ['facecream', 'facewash', 'toothpaste', 'bathingsoap',
       'shampoo', 'moisturizer']
for label in labels:
  plt.plot(data['month_number'],data[label],label=label)

plt.xlabel("Month")
plt.ylabel("Unit number")
plt.legend()
```

### Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot
![img](img/04.png)

```python
plt.scatter(data['month_number'],data['toothpaste'],label='toothpaste sale')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.xlabel("Month")
plt.ylabel("Unit number")
plt.legend()
```

### Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

![img](img/05.png)

```python
X_axis = np.arange(len(data)) 
plt.bar(data['month_number'] - 0.2, data['facecream'], 0.4, label = 'facecream') 
plt.bar(data['month_number'] + 0.2, data['facewash'], 0.4, label = 'facewash')
plt.grid(color = 'gray', linestyle = '--', linewidth = 1)
plt.xlabel("Month")
plt.ylabel("Unit number")
plt.legend()
```

### Exercise 6: Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk

![img](img/06.png)

```python
plt.bar(data['month_number'], data['bathingsoap'], 0.8, label = 'facewash')
plt.grid(color = 'gray', linestyle = '--', linewidth = 1)
plt.xlabel("Month")
plt.legend()
plt.title("Boathing soap")
```
